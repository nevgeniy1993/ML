```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
```

```python
house_df = pd.read_csv('housing.csv')
```

```python
x = house_df.drop('MEDV',axis=1)
y = house_df['MEDV']
```

```python
X_train, X_test, y_train, y_test = train_test_split(x,y, test_size=0.22, random_state=228)
model_lr = LinearRegression()
model_lr.fit(X_train,y_train)
pred = model_lr.predict(X_test)
```

```python
def toFixed(numObj, digits=0):
    return f"{numObj:.{digits}f}"
check_pred=[]
for i in pred:
    check_pred.append(toFixed(i,1))
check = list(zip(check_pred, y_test.tolist()))
check
```

```python
print('Ошибка MSE=',mean_squared_error(y_test,pred,squared=True))

# random state 1922 = Ошибка MSE= 0.722114102636394
```

```python
plt.figure(figsize=(20,9))
plt.plot(y_test.values, "blue", linewidth=5, label = 'Реальные значения')
plt.plot(pred, "red", linewidth=5, label = 'Наш первый ML')
plt.legend(loc='best')
plt.grid(True)
plt.show()
```
